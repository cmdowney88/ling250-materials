## 1. The Overall Idea

Hypothesis testing is the backbone of the **Experimental Scientific Method**.

A **hypothesis** is a claim that can be **supported or refuted by evidence**,
usually from a controlled experiment.

Key principles:

- Experimentalism deals with **evidence** for a claim, rather than proof
- There is **always uncertainty** in the process
- A hypothesis has to be **falsifiable**: if it can't be shown to be wrong,
  it's experimentally useless

In practice, we test new claims against a **null hypothesis** — the hypothesis
that our new claim *isn't* true.

---

## 2. Statistical Hypotheses

To test claims with **data**, we have to translate research hypotheses into
**statistical hypotheses**.

Consider an example:

- **Research hypothesis**: Clairvoyance/ESP is real, allowing people to "see"
  things that aren't physically present
- **Experiment**: 100 people guess the result of a coin flip from a separate
  room
- **Statistical hypothesis**: the proportion of subjects that guess the flip
  correctly will be different from what is expected **by chance**
  ($\theta = 0.5$)

The statistical hypothesis **only** supports the research hypothesis **if the
experiment is well-designed** (set up to support or refute ESP).

---

## 3. The Null Hypothesis

Counterintuitively, we focus on the **negation** of the hypothesis, called
the **Null Hypothesis** or $H_0$.

- The Null Hypothesis often represents a **simpler** or **less-informative**
  state of affairs than the actual hypothesis
  - E.g. $H_0$ for the ESP experiment is "subjects will guess the coin flip
    according to **random chance**"
- The goal of statistical hypothesis testing is to **refute the Null
  Hypothesis**, which is otherwise **assumed to be true**

### "Null until proven otherwise"

The book uses the metaphor of a **legal trial**, where a defendant is presumed
**innocent until proven guilty**.

- The idea: it's **worse to convict an innocent person** than it is to let a
  guilty person go free
- (It's worse to **incorrectly refute the Null Hypothesis**)

Why? This gives us a **rigorous standard of evidence** for new claims.

- Imagine believing everything you heard was true until proven otherwise
- The **onus is on the scientist** to support their claim

### Example: Aliens and the Null Hypothesis

What's the Null Hypothesis when someone says they "saw an alien spacecraft"
or that "aliens built the pyramids"?

- $H_0$: **not aliens**
- This is the fallacy behind shows like *Ancient Aliens* or similar
  "documentaries"

Important: $H_0$ is **NOT** an alternative hypothesis like "you saw an
airplane." It's simply the negation of the claim.

Carl Sagan: **"Extraordinary claims require extraordinary evidence."**

### Null Hypothesis for the ESP Study

**Null Hypothesis** ($H_0$):

- Non-statistically: "ESP will not factor into subjects guessing the coin"
- Statistically: "The subjects will only be able to guess the outcome of the
  coin flip by chance, with a probability of $\theta = 0.5$"

The goal of the test is to **retain** (keep) or **reject** the Null Hypothesis.

**Alternative hypothesis** ($H_1$; what we call the "non-null" hypothesis):

- "Subjects will be able to guess the outcome of the coin flip with a
  probability different from chance: $\theta \neq 0.5$"
- The **experimental design** suggests that this means ESP

---

## 4. Types of Error

Hypothesis testing can result in **two types of error**:

- **Type I**: the null is **incorrectly rejected** (we claim an effect that
  isn't there)
- **Type II**: the null is **incorrectly retained** (we miss a real effect)

The goal is to **minimize Type I errors** — "Null until proven otherwise."

|  | retain $H_0$ | reject $H_0$ |
|:---:|:---:|:---:|
| $H_0$ is true | correct decision | error (Type I) |
| $H_0$ is false | error (Type II) | correct decision |

### Significance and Power

- $\alpha$: probability the test gives a **Type I error** (incorrectly reject
  the null)
  - AKA: the **significance level**
- $\beta$: probability the test gives a **Type II error** (incorrectly retain
  the null)
  - We don't talk about $\beta$ so much as $1 - \beta$, which is called the
    **power** of the test
  - Power is the probability that we **correctly reject the null**
  - Power is closely tied to the **sample size** of the test

**Significance is prioritized over power!**

|  | retain $H_0$ | reject $H_0$ |
|:---:|:---:|:---:|
| $H_0$ is true | $1 - \alpha$ (probability of correct retention) | $\alpha$ (Type I error rate) |
| $H_0$ is false | $\beta$ (Type II error rate) | $1 - \beta$ (power of the test) |

## 5. Elements of a Test

To run a hypothesis test, we need three things:

1. **Test statistic**: the data variable used to support or refute the
   hypothesis
   - i.e. what are we measuring?

2. **Significance level** ($\alpha$): what is the highest Type I error rate
   that we're willing to accept?
   - What chance of incorrectly rejecting $H_0$ are we okay with?

3. **Sampling distribution** of the test statistic: how do we **expect** the
   data to be distributed assuming the null is true?

### ESP Test: Putting It Together

For our ESP experiment:

- **Test statistic**: the number of **correct guesses** out of the total
  trials (say 100 subjects, 1 test each)
- **Significance level**: let's say **0.05 (5%)** to start with
  - This is a little arbitrary, but it's a common threshold in science
- **Sampling distribution**: the number of correct guesses will follow the
  **Binomial distribution**
  - $\theta = 0.5$ if subjects are guessing randomly (Null Hypothesis)

$$X \sim \text{Binomial}(\theta, N)$$

### Sampling Distribution Under the Null

Recall from last time that `dbinom()` gives the probability of each
outcome under the Binomial distribution. Let's plot the sampling
distribution for our test statistic $X$ when the null hypothesis is true
($\theta = 0.5$, $N = 100$):

````r
probs = dbinom(0:100, size = 100, prob = 0.5)
plot(0:100, probs, type = "h", col = "purple", lwd = 2,
     xlab = "Number of Correct Responses (X)", ylab = "Probability",
     main = "Sampling Distribution for X if the Null is True")
````

Since the null hypothesis says that the probability of a correct response
is $\theta = 0.5$, the sampling distribution says that the most likely
value is 50 (out of 100) correct responses. Most of the probability mass
lies between 40 and 60.

### Expected Results

- About how many correct guesses do we expect **assuming the Null is true**?
  - **About 50** — the binomial distribution gives our expected results

- How many correct guesses do we expect **if the Null is false**?
  - Somewhere **not close to 50!**
  - Our alternative hypothesis is **not specific** about whether to expect
    less or more

How much do our results need to **diverge from expectation** in order to
declare them **significant**?

## 6. Critical Regions

**Critical regions** are the values of the test statistic ($X$) which lead us
to **reject the Null**.

- For significance $\alpha = 0.05$, this is the **outer 5%** of the sampling
  distribution
- These values are **possible but unlikely** under the Null
- Thus: $\alpha$ is the chance of **incorrectly rejecting** the Null

### Two-Sided Test

In a **two-sided test**, the critical region is split between both tails
(2.5% on each side for $\alpha = 0.05$). This tests whether the result is
**different from** 50 in **either direction**:

````r
x_vals = 0:100
probs = dbinom(x_vals, size = 100, prob = 0.5)

# Find the critical boundaries (2.5% in each tail)
lower_crit = qbinom(0.025, size = 100, prob = 0.5)
upper_crit = qbinom(0.975, size = 100, prob = 0.5)

# Color the bars: purple for critical region, gray for non-critical
bar_colors = ifelse(x_vals <= lower_crit | x_vals >= upper_crit, "purple", "gray80")

plot(x_vals, probs, type = "h", col = bar_colors, lwd = 2,
     xlab = "Number of Correct Responses (X)", ylab = "Probability",
     main = "Critical Regions for a Two-Sided Test")

# Add labels
text(lower_crit - 8, max(probs) * 0.5, "lower critical region\n(2.5% of the distribution)",
     cex = 0.8)
text(upper_crit + 8, max(probs) * 0.5, "upper critical region\n(2.5% of the distribution)",
     cex = 0.8)
````

The purple bars are the critical regions. If our observed $X$ falls in one
of these regions, we reject $H_0$.

### One-Sided Test

A **one-sided test** has a critical region only to **one side** of the mean.
This represents the hypothesis that ESP **increases** accuracy (rather than
just having an effect):

- $H_0$: $\theta \leq 0.5$
- $H_1$: $\theta > 0.5$

The full 5% goes into one tail instead of being split:

````r
x_vals = 0:100
probs = dbinom(x_vals, size = 100, prob = 0.5)

# One-sided: 5% in the upper tail only
upper_crit_one = qbinom(0.95, size = 100, prob = 0.5)

bar_colors = ifelse(x_vals >= upper_crit_one, "purple", "gray80")

plot(x_vals, probs, type = "h", col = bar_colors, lwd = 2,
     xlab = "Number of Correct Responses (X)", ylab = "Probability",
     main = "Critical Region for a One-Sided Test")

text(upper_crit_one + 10, max(probs) * 0.5, "critical region\n(5% of the distribution)",
     cex = 0.8)
````

The alternative (two-sided) test splits the 5% across both tails. The
one-sided test concentrates it on one side, making it easier to reject
$H_0$ in that direction — but unable to detect an effect in the other
direction.

### Are You Convinced?

Let's say we're doing a **one-sided** test, with a significance threshold
of $\alpha = 0.05$. And let's say that **60 subjects guess correctly** in
the experiment.

This falls in the critical region — it's "**statistically significant**"
given our definitions.

Do you believe in ESP?

````r
x_vals = 0:100
probs = dbinom(x_vals, size = 100, prob = 0.5)

upper_crit_one = qbinom(0.95, size = 100, prob = 0.5)
bar_colors = ifelse(x_vals >= upper_crit_one, "purple", "gray80")

plot(x_vals, probs, type = "h", col = bar_colors, lwd = 2,
     xlab = "Number of Correct Responses (X)", ylab = "Probability",
     main = "Critical Region for a One-Sided Test")

# Mark the observed result
abline(v = 60, col = "red", lwd = 2, lty = 2)
text(65, max(probs) * 0.7, "observed X = 60", col = "red", cex = 0.9)
````

## 7. p-values

The **p-value** is the probability assigned to results that are **at least as
extreme** as the one we have, assuming the Null.

- Essentially: "How probable would this data (or more extreme data) be
  assuming the Null is true?"
- The **more unlikely** a result is assuming the Null, the **more
  significant** it is

Many/most **scientific papers** report the p-value of their result.

**The p-value is NOT the "probability that the Null is true."** It is the
probability of seeing data this extreme *given that* the Null is true.

### Significance Stars

| Usual notation | Signif. stars | English translation |
|:-:|:-:|:--|
| $p > .05$ | | The test wasn't significant |
| $p < .05$ | * | Significant at $\alpha = .05$ but not at $.01$ or $.001$ |
| $p < .01$ | ** | Significant at $\alpha = .05$ and $.01$ but not at $.001$ |
| $p < .001$ | *** | Significant at all levels |

### ESP p-value

We can get the p-value of our ESP experiment by checking the **probability
that the Null assigns** to our result.

R has a built-in function `binom.test()` for the exact binomial test:

````r
# Two-sided binomial test: 60 correct out of 100, null probability = 0.5
binom.test(x = 60, n = 100, p = 0.5)
````

The two-sided p-value is about 0.057 — there is a 5.7% chance of getting
**60 or more** OR **40 or fewer** successes (given the Null). This is
*not* significant at $\alpha = 0.05$.

We can get a **one-sided** test with the `alternative` argument:

````r
# One-sided: test whether theta > 0.5
binom.test(x = 60, n = 100, p = 0.5, alternative = "greater")
````

The one-sided p-value is about 0.028 — significant at $\alpha = 0.05$
but not at $\alpha = 0.01$.

## 8. Power

Recall that **power** ($1 - \beta$) is the probability of **correctly
rejecting** the null when it is false.

Power depends on the **true** value of the parameter. If the true success
rate were $\theta = 0.70$ (i.e., ESP works really well), the sampling
distribution shifts far from 50, and almost all of it falls in the
critical region — high power:

````r
x_vals = 0:100

# Sampling distribution under the alternative: theta = 0.70
probs_alt = dbinom(x_vals, size = 100, prob = 0.70)

# Critical boundaries from the null (two-sided, alpha = 0.05)
lower_crit = qbinom(0.025, size = 100, prob = 0.5)
upper_crit = qbinom(0.975, size = 100, prob = 0.5)

bar_colors = ifelse(x_vals <= lower_crit | x_vals >= upper_crit, "purple", "gray80")

plot(x_vals, probs_alt, type = "h", col = bar_colors, lwd = 2,
     xlab = "Number of Correct Responses (X)", ylab = "Probability",
     main = expression(paste("Sampling Distribution for X if ", theta, " = 0.70")))

text(lower_crit - 8, max(probs_alt) * 0.5,
     "lower critical region\n(2.5% of the distribution)", cex = 0.8)
text(upper_crit + 2, max(probs_alt) * 0.5,
     "upper critical region\n(2.5% of the distribution)", cex = 0.8)
````

Almost all of the distribution lies in the rejection region. If the true
success rate were 0.70, we would almost always correctly reject the null.

But what if the true effect were smaller — say $\theta = 0.55$? The
sampling distribution would overlap heavily with the null distribution,
and power would be low. This is why **larger sample sizes increase power**:
they make the sampling distribution narrower, reducing the overlap.

````r
# Compare the null (theta=0.5) and a subtle alternative (theta=0.55)
x_vals = 0:100
probs_null = dbinom(x_vals, size = 100, prob = 0.5)
probs_alt = dbinom(x_vals, size = 100, prob = 0.55)

plot(x_vals, probs_null, type = "h", col = "gray70", lwd = 2,
     xlab = "Number of Correct Responses (X)", ylab = "Probability",
     main = expression(paste("Null (", theta, "=0.5) vs. Alternative (", theta, "=0.55)")))
lines(x_vals, probs_alt, type = "h", col = rgb(0.5, 0, 0.5, 0.5), lwd = 2)
legend("topright", legend = c(expression(paste(theta, " = 0.5 (Null)")),
                              expression(paste(theta, " = 0.55 (Alternative)"))),
       col = c("gray70", rgb(0.5, 0, 0.5, 0.5)), lwd = 2)
````

When the true effect is small (e.g., $\theta = 0.55$), the two
distributions overlap substantially. Most samples from the alternative
will *not* fall in the critical region, so we'll often fail to reject the
null even though it's false — that's low power.

## Quick Reference

| Concept | Definition |
|---------|------------|
| Hypothesis | A falsifiable claim that can be supported or refuted by evidence |
| Null hypothesis ($H_0$) | The "default" — assumes no effect or no difference |
| Alternative hypothesis ($H_1$) | The claim we're trying to find evidence for |
| Type I error | Incorrectly rejecting $H_0$ (false positive) |
| Type II error | Incorrectly retaining $H_0$ (false negative) |
| Significance level ($\alpha$) | Maximum acceptable Type I error rate |
| Power ($1 - \beta$) | Probability of correctly rejecting $H_0$ when it's false |
| p-value | Probability of data at least as extreme as observed, assuming $H_0$ |
| Critical region | Values of the test statistic that lead to rejecting $H_0$ |
| One-sided test | Critical region on one side only (tests a directional claim) |
| Two-sided test | Critical region on both sides (tests for any difference) |

| Task | R code |
|------|--------|
| Binomial test | `binom.test(x, n, p)` |
| One-sided binomial test | `binom.test(x, n, p, alternative = "greater")` |
| Binomial probability | `dbinom(x, size, prob)` |
| Binomial quantile | `qbinom(p, size, prob)` |
