## 2. R Language Basics

### Similarities to Python

**Arithmetic operators** work the same way:

````r
# Addition, subtraction
3 + 5
10 - 4

# Multiplication, division, exponentiation
6 * 7
20 / 4
2 ^ 8

# Comparisons
10 > 5
3 == 3
4 != 7
````

**Strings** are written in double quotes:

````r
my_string = "This is a string"
my_string
````

Note: In R, strings and individual characters are the same type, both called
"character" (unlike Python, which distinguishes `str` from `chr`). String
manipulation functions exist but are less intuitive than Python's.

**Comments** use `#`:

````r
# This is a comment — R ignores it
x = 42  # inline comments work too
````

### Minor Syntax Differences

**Variable assignment** — R has two equivalent options:

````r
# Both of these do the same thing
my_variable = 10
my_variable <- 10

my_variable
````

The arrow operator (`<-`) is the traditional R style; `=` was added later and
is now widely accepted. We'll use `=` in this course.

**Boolean operators** — same logic as Python, different symbols:

````r
# "and": & (not "and")
# "or": | (not "or")
# "not": ! (not "not")
# Boolean values: TRUE / FALSE (not True / False)

TRUE & FALSE
TRUE | FALSE
!TRUE

# Parentheses needed for order of operations
(2 > 1) & (3 > 2)
````

**Other constructs** (functions, if/else, loops) work similarly to Python but
with syntactic differences. We won't focus on these heavily — see *Learning
Statistics with R* Ch. 3 if you need them.

### Vectors: R's Version of Python Lists

The most important data structure difference between R and Python is how
sequences work. In Python, you use lists. In R, the equivalent is a **vector**.

````r
# Create a vector with the c() function ("combine")
my_vec = c(1, 2, 3, 4, 5)
my_vec
````

Key differences from Python lists:

**1. Indexing starts at 1 (not 0):**

````r
my_vec[1]   # first element is 1, not 0
my_vec[5]   # last element
````

**2. Elements can be reassigned:**

````r
my_vec[1] = 0
my_vec
````

**3. Arithmetic applies to all elements at once (vectorized):**

````r
my_vec = c(1, 2, 3, 4, 5)  # reset

my_vec * 2         # multiply every element by 2
my_vec + c(1, 2, 3, 4, 5)  # element-wise addition of two vectors
````

**4. `length()` gives the number of elements:**

````r
length(my_vec)
````

### Filtering Data Frames with `filter()`

`filter()` returns a **subset** of a data frame where certain conditions hold:

````r
# All rows where the vowel is "i"
filter(vowels, VOWEL == "i")
````

`filter()` can take **any number of conditions**, all of which must be true:

````r
# Female speakers producing the vowels "i" or "u",
# whose height is at or below the mean height of all speakers
filter(
  vowels,
  SEX == "female",
  VOWEL == "i" | VOWEL == "u",
  HEIGHT <= mean(HEIGHT)
)
````

Note: `mean(HEIGHT)` here refers to the HEIGHT column of the data frame —
dplyr makes columns available by name within filter conditions.

## 4. Variable Types and Scales of Measurement

### What Is a Variable?

A **variable** is a measurable attribute of data — something you record for
each observation. Variables come from:

- **Instrumentation or observation** (e.g., vowel formants measured with a
  microphone and audio software)
- **Human responses** (e.g., a survey answer)

Variables are usually stored in **columns** of a data frame or CSV. Critically,
variables are **not all on the same scale** — you can't directly compare birth
year, eye color, and vowel formants.

### Four Scales of Measurement

Our textbook (LSR) defines four scales. The scale tells you **how you should
analyze** a variable.

#### Nominal (Categorical)

Nominal variables have **no meaningful ordering or relationship** between values.
The only thing you can say is that values are different.

- Examples: eye color, sex, citizenship, language spoken, transportation method
- Neither French nor German is "greater than" the other, and they **can't be
  averaged**
- I will usually call these **categorical variables**

#### Ordinal

Ordinal variables have a **natural, meaningful ordering** between values, but
**no other structure**.

- Examples: position in a race, ranked choice voting, other rankings
- Finishing 1st vs. 2nd tells you who was faster, but **not by how much**
- Addition and subtraction are **not well-defined** (1st place isn't "3rd minus
  2nd")
- Ordinal values also **can't be meaningfully averaged**

#### Interval

Interval variables are **ordered** and **differences are meaningful**, but they
have **no true zero value**.

- Examples: Fahrenheit and Celsius, calendar year
- The difference between 60° and 65° is the same as between 20° and 25°
- But: 0°C is not "no heat," and 40° is not "twice as hot" as 20°
- **Not meaningful to multiply or divide**

#### Ratio

Ratio variables have **ordering**, **meaningful differences**, and a **true
zero**.

- Examples: race finish time, weight, age, sports scores
- A 30 y/o is 5 years older than a 25 y/o (30 − 25), or 1.2× older (30 / 25)
- Ratio variables **can be meaningfully averaged**
- These are what you might think of as "regular" numbers

### Continuous vs. Discrete

- **Continuous**: for any two values, it's possible to have a value in-between
  (e.g., between 2.0 and 3.0 is 2.5, then 2.05, etc.)
- **Discrete**: not continuous (e.g., the number that comes up after rolling a die)

| | continuous | discrete |
|---------|:---:|:---:|
| nominal | | ✓ |
| ordinal | | ✓ |
| interval | ✓ | ✓ |
| ratio | ✓ | ✓ |

*From LSR p.17*

Nominal and ordinal variables are always discrete. Interval and ratio variables
can be either.

### Factors in R

Sometimes a **categorical variable is coded as a number** in a CSV — for
example, a "sex" column might use 0 for male, 1 for female. R will naively
treat these as numeric, which can cause problems:

````r
sex_numeric = c(0, 1, 0, 0, 1, 0, 0, 0)

# R treats this as a number — inappropriate operations succeed silently
sex_numeric < 2       # all TRUE (meaningless)
sex_numeric / 8       # silently produces garbage
````

The R data type **factor** designates a categorical variable and prevents
inappropriate manipulation:

````r
sex_factor = as.factor(sex_numeric)
sex_factor            # shows levels

# Now arithmetic raises a warning and returns NA
sex_factor + 1
````

You can give the **levels** meaningful names with `levels()`:

````r
sex_named = as.factor(c(0, 1, 0, 0, 1, 0, 0, 0))
levels(sex_named) = c("male", "female")
sex_named
````

This also prevents accidental equivalence between two different factor
variables that happen to use the same numbers — after naming, R will error if
you try to compare factors with different level sets.

## 5. Descriptive Statistics

**Descriptive statistics** aim to **concisely and meaningfully summarize** data.
(As opposed to *inferential* statistics, which make inferences from data — more
on that in the coming weeks.)

Two broad categories:

- **Central tendency**: where the "middle" of the data is (mean, median, mode)
- **Variability**: how much the data deviates from the central tendency
  (variance, standard deviation)

### Central Tendency

#### Mean

The **mean** (what most people call the "average") is the sum of all data
points divided by the total number of data points:

$$\bar{X} = \frac{1}{N} \sum_{i=1}^{N} X_i$$

````r
mean(c(1, 2, 3, 4, 5))

# Mean F1 in the vowels dataset
mean(vowels$F1)
````

#### Median

The **median** is the middle datapoint when values are arranged in order. It is
equivalent to the **50th percentile**: half the data falls above it, half below.

````r
# 3 is the median of both of these:
median(c(1, 2, 3, 4, 5))
median(c(1, 2, 3, 8, 10))   # notice mean would differ, but median is same

# For an even number of points, median = average of the middle two
median(c(1, 2, 3, 4))       # (2 + 3) / 2 = 2.5

# Median height in the vowels dataset
median(vowels$HEIGHT)
````

#### Mean vs. Median

**Advantages of the mean:**

- Takes every datapoint into account (median ignores all but the middle)
- Mathematically important — used in many statistical formulas
- Almost everyone understands it as the "average"

**Advantages of the median:**

- **Resistant to outlier datapoints** — this is why people report "median
  salary" or "median home price": a handful of ultra-high earners skew the
  mean upward
- Appropriate for **ordinal data**, whereas the mean is not

````r
# Simulated salaries: most people earn ~50k, one billionaire
salaries = c(45000, 48000, 52000, 55000, 60000, 1000000000)
mean(salaries)
median(salaries)
````

### Variability

#### Variance

**Variance** is the mean squared deviation from the mean:

$$Var(X) = \frac{1}{N} \sum_{i=1}^{N} (X_i - \bar{X})^2$$

Squaring serves the same role as absolute value — it makes all deviations
positive. Variance is mathematically important, but the units are squared
(e.g., Hz²), which makes it hard to interpret directly.
Just like variance.

````r
var(vowels$F1)
````

#### Standard Deviation

**Standard deviation** is the square root of the variance:

$$StDev(X) = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (X_i - \bar{X})^2}$$

Often written as σ (sigma). Because it's in the **same units as the original
variable**, it's easier to interpret than variance. You'll often see a
distribution described as $\bar{X} \pm StDev(X)$.


````r
sd(vowels$F1)

# Report mean ± SD
cat("F1:", round(mean(vowels$F1), 1), "±", round(sd(vowels$F1), 1), "Hz\n")
````

#### Quantiles and Percentiles

A **quantile** (or **percentile**) is a value below which a certain percentage
of the data falls. The median is the 50th percentile.

````r
# 50th percentile of height (should equal the median)
quantile(vowels$HEIGHT, probs = 0.5)

# 25th, 50th, and 75th percentiles
quantile(vowels$HEIGHT, probs = c(0.25, 0.5, 0.75))
````

#### Interquartile Range (IQR)

The **IQR** is the difference between the 75th and 25th percentiles — it shows
how wide the **middle 50%** of the data is. Useful compared to the total range:

````r
IQR(vowels$HEIGHT)

# Compare to total range
total_range = max(vowels$HEIGHT) - min(vowels$HEIGHT)
total_range
````

The IQR is 12.25 cm while the total range is 36 cm — the middle half of the
data is relatively compact compared to the full spread.

### Summary Statistics

R's `summary()` function gives several statistics in one call, and works on
either a vector or a full data frame:

````r
summary(vowels$HEIGHT)
````

````r
summary(vowels)
````

Note that for character columns (SPEAKER, WORD, VOWEL, SEX), `summary()` just
reports the class and length — it can only compute numeric statistics for
numeric columns.

## 6. Basic Plotting

R has built-in functions for quick data visualization. These are great for
**exploratory analysis** — getting a fast look at your data. For
publication-quality figures, we'll use `ggplot2` (coming in a few weeks).

### Scatterplots with `plot()`

`plot()` with **one vector** plots the values on the y-axis (by index). With
**two vectors**, it makes a **scatterplot**:

````r
# Scatterplot of F1 vs. F2
plot(vowels$F1, vowels$F2)
````

Notice the negative correlation — higher F1 tends to go with lower F2, which
corresponds to the vowel space in phonetics.

### Histograms with `hist()`

`hist()` plots the **frequency** of values in certain ranges, giving you a
sense of the data's distribution:

````r
hist(vowels$HEIGHT)
````

The `breaks` argument controls approximately how many bins to use:

````r
# Fewer bins — broader picture
hist(vowels$HEIGHT, breaks = 4)

# More bins — finer detail
hist(vowels$HEIGHT, breaks = 16)
````

### Boxplots with `boxplot()`

`boxplot()` shows the **inner quartiles** of the data. The box covers the
middle 50% (from 25th to 75th percentile), the line inside is the median, and
the whiskers extend to the data's range (excluding outliers):

````r
boxplot(vowels$HEIGHT)
````

Note that `boxplot()` doesn't add axis titles by default. You can add them:

````r
boxplot(vowels$HEIGHT, ylab = "Height (cm)", main = "Speaker Heights")
````
